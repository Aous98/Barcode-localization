```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
file = pd.read_csv('/home/aous/Desktop/MIPT/project/full data resized/predict/prediction.csv')
file.head()
```

```python
pred_file = file.copy()

pred_file
file['File Name'] = file['File Name'].str.rstrip('.json')

file.head()
```

```python
file.fillna(0, inplace=True)
```

```python
if file.isna().values.any():
    print("The CSV file contains NaN values.")
else:
    print("The CSV file does not contain any NaN values.")
```

```python
import os 
st = 'File Name'
saving_dir = '/home/aous/Desktop/MIPT/project/full data resized/predict/temp mask/'
for i in range(len(file)):
    width, height = file['Value 1'][i], file['Value 2'][i]
    black_image = np.full((height, width), 0, dtype=np.uint8)
    points = [
    (file['Value 3'][i], file['Value 4'][i]),
    (file['Value 5'][i], file['Value 6'][i]),
    (file['Value 7'][i], file['Value 8'][i]),
    (file['Value 9'][i], file['Value 10'][i]),
    (file['Value 11'][i], file['Value 12'][i]),
    (file['Value 13'][i], file['Value 14'][i])
    ]
    polygon = [np.array(points)]
    
    cv2.fillPoly(black_image, np.int32(polygon), color=255)
    mask_path = saving_dir + f'{file[st][i]}.png'
    cv2.imwrite(mask_path ,black_image)
```
